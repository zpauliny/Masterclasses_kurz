# **Častice**

Častice sú základné jednotky hmoty, ktorá tvorí vesmír. V predchádzajúcej kapitole sme už spoznali nejaké častice - protón, neutrón a elektrón. Z nich iba elektrón je elementárnou časticou, protón a neutrón nie sú elementárne, pretože sa skladajú z iných častíc. Častice si preto rozdelíme na dve veľké skupiny podľa toho, či majú alebo nemajú nejakú vnútornú štruktúru: na **elementárne** a **zložené** častice.

## **1. Elementárne častice**

## **Prenášači fundamentálnych síl**

**Gluóny, fotóny, bozóny Z$^0$ a W$^\pm$** sú prenášačmi fundamentálnych síl (napr. elektromagnetická sila, slabá a silná jadrová sila) a sú tiež elementárnymi časticami. Kým gluóny a fotóny majú nulovú hmotnosť, bozóny Z a W sú veľmi tažké. O interakciach, ktoré sprostredkúvajú, si povieme bližšie v nasledujúcej kapitole.

Ďalším bozónom, ktorý nemá vnútornú štruktúru a má špeciálnu funkciu, je **Higgsov bozón**. Ten je poslednou objavenou elementárnou časticou, bol potvrdený v roku 2012. Viac o ňom v ďalšej kapitole.

## **2. Zložené častice**

sú častice zložené z viacerých elementárnych častíc. Sú to najmä **hadróny**, ktoré sú tvorené kvarkami. Kvarky majú totiž zváštnu vlastnosť, že sa v prírode nevyskytujú samostatne, vždy sa spoja do väčších častíc. Existujú dva typy hadrónov:

- **Mezóny**: Hadróny tvorené jedným kvarkom a jedným antikvarkom (antičastica kvarku, vysvetlíme nižšie), napríklad pión, kaón.
- **Baryóny**: Hadróny tvorené z troch kvarkov, napríklad protón, neutrón, lambda.

## **Antičastice**

Poznáme aj antičastice, ktoré sú „zrkadlovými náprotivkami“ častíc. Každá častica má svoju zodpovedajúcu antičasticu, ktorá má **rovnakú hmotnosť**, ale **opačný elektrický náboj**. 

**Napríklad:**

- **Elektrón (e⁻)** má antičasticu **pozitrón (e⁺)**, ktorý má kladný náboj.
- **Protón (p⁺)** má **antiprotón (p⁻)** s negatívnym nábojom.
- **Neutrón (n)** má **antineutrón**, ktorý je elektricky neutrálny, ale je zložený z **antikvarkov**.

Keď sa častica stretne so svojou antičasticou, dôjde k **anihilácii** – premenia sa na energiu vo forme **gama žiarenia (fotónov)**.

Niektoré častice sú **samy sebe antičasticou**. To znamená, že nemajú samostatnú "anti-verziu", pretože už v základnej forme majú vlastnosti, ktoré by inak mali ich antičastice.

**Napríklad:**

- **Fotón (γ)** – Nemá elektrický náboj, takže nemá odlišnú antičasticu.
- **Z-bozón (Z)** – Neutrálna intermediálna častica pri slabých interakciách.
